```python
import cv2 
import mediapipe as mp #mediapie libraies&solutions
import numpy as np #mutidimensional array object
mp_drawing = mp.solutions.drawing_utils #drawing utilities & visualizing pose
mp_pose = mp.solutions.pose #pose estimation models
```

```python
#DETECTIONS
cap = cv2.VideoCapture(0) #setting video cam devices #SETUP MIDIAPIPE INSTANCE
with mp_pose.Pose(min_detection_confidence=0.5,min_tracking_confidence=0.5) as pose: #detection accuracy
    while cap.isOpened(): #loop 
        ret,frame = cap.read() #store & extract from cap,return to ret, frame gives image captured
        cv2.imshow("Mediapipe Feed", frame) #display

        #recolor image to RGB
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB) #reorder color range
        image.flags.writeable = False #set to unwriteable
        #opencv image format: BGR

        #make detection
        results = pose.process(image)

        #recolor image back to BGR
        image.flags.writeable = True
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        
        #render detections
        mp_drawing.draw_landmarks(image,results.pose_landmarks,mp_pose.POSE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color=(245,117,60),thickness=2,circle_radius=2), #joint
                                 mp_drawing.DrawingSpec(color=(245,66,230),thickness=2,circle_radius=2), #connection between bones
                                 )
        
        
        cv2.imshow('Mediapipe Feed', image)

        if cv2.waitKey(10) & 0xFF == ord('q'): #close dislay
            break

cap.release() #release web cam
cv2.destroyAllWindows() #close feed
```
